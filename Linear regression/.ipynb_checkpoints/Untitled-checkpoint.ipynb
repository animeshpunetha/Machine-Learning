```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df
```

```python
plt.scatter(df.area, df.price, color = 'red', marker = 'x')
```

# now get the linear regression equeation using sklearn

```python
# Create linear regression object
newdf = df.drop('price', axis='columns')
newdf
```

```python
price = df.price
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(newdf, price)
```

```python
reg.predict([[5000]])
```

## Generate CSV file with list of home price predictions

```python
df2 = pd.read_csv('areas.csv')
df2
```

```python
ans = reg.predict(df2)
ans
```

```python
df2['prices'] = ans
df2
```

```python
df2.to_csv('generated_output.csv', index = False)
```

```python
p = pd.read_csv('generated_output.csv')
p
```
